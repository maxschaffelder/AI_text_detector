```python
import pandas as pd
import re

df = pd.read_csv('/Users/maxschaffelder/Desktop/Thesis/Data/data_standpoint_1.csv') 
df.head()
```

```python
# Define a function to remove references from the text
def remove_references(text):
    # Remove references by matching content within round brackets and replacing them with an empty string
    return re.sub(r'\(.*?\)', '', text)

# Apply the function to the second column 
df['content'] = df['content'].apply(remove_references)

```

```python
# Save the modified DataFrame to a new CSV file

output_csv = '/Users/maxschaffelder/Desktop/Thesis/Data/data_standpoint_2.csv'
df.to_csv(output_csv, index=False)
```

```python
import re

# Finding s-numbers in txt
def contains_s_and_seven_digits(text):
    # Define the regular expression pattern
    pattern = r'[Ss]\d{7}'
    
    # Search for the pattern in the text
    match = re.search(pattern, text)
    
    # Return True if the pattern is found, otherwise False
    return bool(match)

for i in range(len(df)):
    
    if contains_s_and_seven_digits(df.iloc[i]["content"]):
        print(i, df.iloc[i]["filename"])

```
